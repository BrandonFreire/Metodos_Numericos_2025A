# Método de la aproximación falsa

Maneja las mismas condiciones que el metodo de la secante pero agrega una condición adicional que permite que se llegue a la solucion. 

**Metodos cerrados:** Obligan que la solucion se encuentre en el intervalo $[a,b]$. Son los metodos de la bisección y el de la aproximación falsa.

La diferencia entre el metodo de la secante y el de la aproximacion falsa es que tienen un punto en comun. Pero este punto no siempre es el mismo.


Se basa en el hecho de tomar puntos que tienen signos diferentes 

**¿Es mejor el metodo de la aproximación falsa que el de la secante?**

En terminos de velocidad de convergencia, el metodo de la secante va a converger mas rapido. Ya que el metodo de la aproximacion falsa va a ser un poco mas lento. Pero el método de la aproximacion falsa es más seguro, porque siempre converge. 


Sea $f(x) = cos(x) - x = 0$

Donde $p0 = 0.5$ y $p1 = pi/4$ y se tiene una tolerancia de $10^{-6}$

A continuación se muestra el código para la implementación del método de la aproximación falsa. 

```python
import math

def f(x):
    return math.cos(x) - x  # Cambia la función según tu problema

def posicion_falsa(xn_2, xn_1, tol, iter):
    i = 2
    f_xn_2 = f(xn_2)
    f_xn_1 = f(xn_1)
    while i <= iter:
        if f_xn_1 - f_xn_2 == 0:
            print("División por cero detectada. El método falla.")
            return None
        
        x_n = xn_1 - f_xn_1 * (xn_1 - xn_2) / (f_xn_1 - f_xn_2)

        print(f"Iteración {i-1}: xn_2 = {xn_2}, xn_1 = {xn_1}, x_n = {x_n}, f(xn_2) = {f(xn_2)}, f(x_n) = {f(x_n)}")

        if abs(x_n - xn_1) < tol:
            print(f"El procedimiento fue exitoso. \n Raíz aproximada: {x_n} \n Iteraciones: {i-1}")
            return
        i += 1
        q = f(x_n)
        if q * f_xn_1 < 0:
            xn_2 = xn_1
            f_xn_2 = f_xn_1
        xn_1 = x_n
        f_xn_1 = q
    print(f"El método falló después de {iter} iteraciones.")
    return None

# Ejemplo de uso con la función coseno analizada anteriormente en los métodos de Newton y bisección:
xn_2 = 0.5      # Primer valor inicial
xn_1 = math.pi/4    # Segundo valor inicial
tol = 1e-6    # Tolerancia
iter = 100      # Número máximo de iteraciones

posicion_falsa(xn_2, xn_1, tol, iter)
```
