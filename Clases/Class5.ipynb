# Método del punto fijo

Un punto fijo para una función es un número en el que el valor de la función no cambia cuando se apllica la función. 

$p = g(p)$

donde p es igual a una funcion g dentro de ese punto p.

Justo en el punto p, la función g va a tomar ese mismo valor.

**¿Que debe haber para que exista un punto fijo?**

- La función g debe ser continua en el intervalo [a,b].

- La derivada de g en el intervalo [a,b] debe ser menor que 1. $|g'(x)| < 1$

## Procedimiento:

- Despejando la variable x de la ecuación f(x) = 0.

# Método de Newton-Raphson

El método de Newton-Raphson es un método iterativo para encontrar raíces de funciones. Se basa en la idea de aproximar la función por su tangente en un punto inicial y luego encontrar el punto donde esa tangente cruza el eje x.

**¿Que es la derivada?**

Es la pendiente de la tangente a la curva de la función original.

### Entradas del seudocódigo del método de Newton-Raphson:	

**¿Que es p0 o el valor de aproximación inicial?**

*P0, P1 y P2 son las aproximaciones a P.* 

- La aproximación incial es importante porque ayuda a que el algoritmo converja lo más rápido al valor real. Si la aproximación inicial está muy lejos de la raíz, el algoritmo puede no converger o converger a una raíz diferente.

- La tolerancia ayuda a determinar cuándo se ha alcanzado una solución suficientemente precisa. Si la diferencia entre dos aproximaciones consecutivas es menor que la tolerancia, el algoritmo se detiene y considera que ha encontrado una raíz.

- El número de iteraciones es el número de veces que se aplica el método de Newton-Raphson para encontrar la raíz. Si el número de iteraciones es muy alto, puede ser un indicativo de que la aproximación inicial no es adecuada o que la función tiene un comportamiento complicado cerca de la raíz.

### Problemas del metodo de newton-raphson 

**¿Que sucede cuando la ecuacion no tiene raiz?**

El metodo no va a converger nunca, se va a acercar al valor y se va a alejar.

**¿Como se dan cuenta de que una ecuación no tiene raíz?**

El error va a dar un indicio de que la ecuacion no tiene raíz. En cuanto mas se aproximan el error va a disminuir, pero si el error aumenta, es porque la ecuacion no tiene raíz. Por eso importante fijarse en el error que se va calculando en cada interación. 

-> En el método de la biseccion es en el extremo negativo en donde se va acercando a la solución.

A continuación de implementa en código el método de Newton-Raphson.

```python
import math

# Defnición de la función
def f(x):
    return math.cos(x) - x

# Definición de la derivada de la función
def df(x):
    return -math.sin(x) - 1

# Método de Newton-Raphson
def newton_raphson(xn_1, tol, iter):
    i = 1
    while i <= iter:
        x_n = xn_1 - f(xn_1)/df(xn_1)

        print(f"Iteración {i}: xn_1 = {xn_1}, x_n = {x_n}, f(xn_1) = {f(xn_1)}, f'(x_n) = {df(xn_1)}")

        if abs(x_n - xn_1) < tol:
            print(f"El procedimiento fue exitoso. \n Raíz aproximada: {x_n} \n Iteraciones: {i}")
            return x_n
        i += 1
        xn_1 = x_n
    print(f"El método falló después de {iter} iteraciones.")
    return None

# Ejemplo de uso con la función coseno analizada anteriormente:
xn_1 = math.pi/4  # Aproximación inicial
tol = 1e-6      # Tolerancia
iter = 100        # Número máximo de iteraciones

newton_raphson(xn_1, tol, iter)
```

## Método de la secante

Tiene como objetivo encontrar raices. Ya no tiene derivadas. 

**¿Que es una raiz?**

Es un valor que hace que la función sea cero.

Se tiene que los dos puntos se van acercando y con ello se va a encontrar la raíz al acercar a una especie de derivada.

**¿Por que la suma y la resta tiene menos errores que la multiplicación y la división?**

Porque la suma y la resta son operaciones más simples y menos propensas a errores de redondeo. La multiplicación y la división pueden introducir errores de redondeo más significativos, especialmente cuando se trabaja con números muy grandes o muy pequeños.

Siendo la pendiente está sección:



Este metodo tiene dos valores de aproximación. Y va a empezar desde 2.

-> Deber 0.5 y 1 para el viernes.
-> Notebook en metodo de newton tomar en cuenta el punto fijo.

A continuación se implementa el método de la secante:

```python
def secante(xn_2, xn_1, tol, iter):
    i = 2
    while i <= iter:
        if f(xn_1) - f(xn_2) == 0:
            print("División por cero detectada. El método falla.")
            return None
        x_n = xn_1 - f(xn_1)*(xn_1 - xn_2)/(f(xn_1) - f(xn_2))

        print(f"Iteración {i-1}: xn_2 = {xn_2}, xn_1 = {xn_1}, x_n = {x_n}, f(xn_2) = {f(xn_2)}, f(xn_1) = {f(xn_1)}")

        if abs(x_n - xn_1) < tol:
            print(f"El procedimiento fue exitoso. \n Raíz aproximada: {x_n} \n Iteraciones: {i-1}")
            return x_n
        xn_2, xn_1 = xn_1, x_n
        i += 1
    print(f"El método falló después de {iter} iteraciones.")
    return None

# Ejemplo de uso con la función coseno analizada anteriormente en el método de Newton-Raphson:
xn_2 = 0.5      # Primer valor inicial
xn_1 = math.pi/4      # Segundo valor inicial
tol = 1e-6    # Tolerancia
iter = 100      # Número máximo de iteraciones

secante(xn_2, xn_1, tol, iter)
```
